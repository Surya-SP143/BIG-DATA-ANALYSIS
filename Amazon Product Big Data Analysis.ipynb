```python
# The path to the csv
file_path = "/Volumes/workspace/default/my_volume/amazon.csv"

# Loading the data into a Spark DataFrame
df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(file_path)

# Show the first 5 rows to confirm it loaded
display(df.limit(5))
```

```python
from pyspark.sql.functions import col, regexp_replace

# 1. Clean the price column (remove '$' and commas)
df_cleaned = df.withColumn("price_numeric", 
    regexp_replace(col("price"), "[$,]", "").cast("float"))

# 2. Remove rows where the price is missing
df_final = df_cleaned.na.drop(subset=["price_numeric"])

# 3. Rename columns for better readability if needed
df_final = df_final.withColumnRenamed("product_name", "Product")

print("Data Cleaning Complete!")
```

```python
from pyspark.sql.functions import col, regexp_replace

# 1. Clean the price column: Remove symbols and commas, then convert to a number (float)
# Note: Check your table; if your column is named 'discounted_price', use that name below
df_cleaned = df.withColumn("price_numeric", 
    regexp_replace(col("discounted_price"), "[₹,]", "").cast("float"))

# 2. Drop rows where the price is now null (to keep the data clean)
df_final = df_cleaned.na.drop(subset=["price_numeric"])

 # 3. Rename columns for better readability if needed
df_final = df_final.withColumnRenamed("product_name", "Product")


print("Step 2 Complete: Data is cleaned and prices are now numeric!")
```

```python
from pyspark.sql.functions import col, regexp_replace

# 1. Clean the column by removing symbols
# 2. Use .cast("float") but Spark will now handle errors gracefully
df_cleaned = df.withColumn("actual_price_numeric", 
    regexp_replace(col("actual_price"), "[₹,$,]", "").cast("float"))

# 3. Filter out the rows that failed to convert (like '60Hz Refresh Rate')
df_final = df_cleaned.filter(col("actual_price_numeric").isNotNull())

print("Cleaning Complete! Weird text values have been filtered out.")
```

```python
from pyspark.sql.functions import col, regexp_replace, expr

# 1. Clean out the currency symbols and commas
# 2. Use 'try_cast' so weird text doesn't crash the notebook
df_cleaned = df.withColumn("actual_price_numeric", 
    expr("try_cast(regexp_replace(actual_price, '[₹,$,]', '') AS FLOAT)"))

# 3. Filter out those rows that weren't numbers (like '60Hz')
df_final = df_cleaned.filter(col("actual_price_numeric").isNotNull())

print("Cleaned successfully! Malformed rows have been skipped.")
```

```python
# Grouping by category and calculating average price
avg_price_by_category = df_final.groupBy("category") \
    .avg("actual_price_numeric") \
    .orderBy("avg(actual_price_numeric)", ascending=False)

# Display your top 5
display(avg_price_by_category.limit(5))
```

```python
from pyspark.sql.functions import when

# 1. Convert rating to numeric
df_transformed = df_final.withColumn("rating_numeric", col("rating").cast("float"))

# 2. Create a Sentiment Label based on the rating
df_transformed = df_transformed.withColumn("customer_satisfaction", 
    when(col("rating_numeric") >= 4.5, "Excellent")
    .when((col("rating_numeric") >= 4.0) & (col("rating_numeric") < 4.5), "Good")
    .otherwise("Average"))

display(df_transformed.select("product_name", "rating_numeric", "customer_satisfaction").limit(10))
```

```python
# First, ensure discounted_price is also numeric (reuse your cleaning logic)
df_transformed = df_transformed.withColumn("discounted_price_numeric", 
    expr("try_cast(regexp_replace(discounted_price, '[₹,$,]', '') AS FLOAT)"))

# Calculate Savings Amount and Discount Percentage
df_transformed = df_transformed.withColumn("savings_amount", 
    col("actual_price_numeric") - col("discounted_price_numeric"))

df_transformed = df_transformed.withColumn("discount_percentage", 
    (col("savings_amount") / col("actual_price_numeric")) * 100)

display(df_transformed.select("product_name", "actual_price_numeric", "discounted_price_numeric", "discount_percentage").limit(5))
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import rank

# Create a 'Window' to look at each category separately
windowSpec = Window.partitionBy("category").orderBy(col("actual_price_numeric").desc())

# Rank products within their category
ranked_df = df_transformed.withColumn("rank", rank().over(windowSpec))

# Filter to show only the #1 most expensive product per category
top_products_per_cat = ranked_df.filter(col("rank") == 1)

display(top_products_per_cat.select("category", "product_name", "actual_price_numeric"))
```

```python
from pyspark.sql.functions import avg, when

# 1. Calculate the overall average price of the whole dataset
overall_avg = df_final.select(avg("actual_price_numeric")).collect()[0][0]

# 2. Flag products that are significantly more expensive than average
df_final = df_final.withColumn("price_segment", 
    when(col("actual_price_numeric") > overall_avg * 1.5, "Premium")
    .when(col("actual_price_numeric") < overall_avg * 0.5, "Budget")
    .otherwise("Mid-Range"))

display(df_final.select("product_name", "actual_price_numeric", "price_segment").limit(10))
```

```python
from pyspark.sql.functions import split

# Take only the first 5 words of the product name to make it readable
df_final = df_final.withColumn("short_name", 
    expr("concat_ws(' ', slice(split(product_name, ' '), 1, 5))"))

display(df_final.select("short_name", "category", "price_segment").limit(10))
```

```python
from pyspark.sql.functions import split, count

# 1. Extract the first word from the product name as the Brand
df_final = df_final.withColumn("brand", split(col("product_name"), " ")[0])

# 2. Count products per brand to find market leaders
brand_share = df_final.groupBy("brand").count().orderBy("count", ascending=False)

display(brand_share.limit(10))
```

```python
from pyspark.sql.functions import col

# 1. Convert the 'rating' column to a float (number)
# 2. We use 'try_cast' again just in case there is weird text in the ratings
df_final = df_final.withColumn("rating_numeric", col("rating").cast("float"))

# 3. Drop rows where rating is missing so the math doesn't fail
df_final = df_final.dropna(subset=["rating_numeric"])

print("Success: 'rating_numeric' column created and ready for analysis!")
```

```python
from pyspark.sql.functions import col, expr

# 1. We use try_cast to turn the rating into a float
# 2. If it sees a '|' or any text, it will safely turn it into NULL instead of crashing
df_final = df_final.withColumn("rating_numeric", expr("try_cast(rating AS FLOAT)"))

# 3. Filter out the NULLs (the rows that had '|' or missing data)
df_final = df_final.filter(col("rating_numeric").isNotNull())

print("Ratings cleaned! All symbols like '|' have been removed.")
```

```python
from pyspark.sql.functions import avg, stddev

# Now this will work because 'rating_numeric' exists!
stats_df = df_final.groupBy("category").agg(
    avg("rating_numeric").alias("cat_avg_rating"),
    stddev("rating_numeric").alias("cat_stddev_rating")
)

df_stats = df_final.join(stats_df, "category")
df_stats = df_stats.withColumn("rating_performance", 
    (col("rating_numeric") - col("cat_avg_rating")) / col("cat_stddev_rating"))

display(df_stats.select("product_name", "category", "rating_performance").limit(10))
```

```python
# Create a 'Value Score'
# Formula: Rating / Price (higher is better value)
df_final = df_final.withColumn("value_score", col("rating_numeric") / col("actual_price_numeric"))

# Find the top 'Value' products in the dataset
top_value_products = df_final.orderBy("value_score", ascending=False)

display(top_value_products.select("product_name", "actual_price_numeric", "rating_numeric", "value_score").limit(10))
```

```python
from pyspark.sql.functions import avg, col, expr

# 1. Use 'try_divide' to handle the math safely. 
# 2. If a category has only 1 item, it will return NULL instead of crashing.
df_stats = df_stats.withColumn("rating_performance", 
    expr("try_divide((rating_numeric - cat_avg_rating), cat_stddev_rating)"))

# 3. Create the summary, ignoring those single-item categories
final_summary = df_stats.groupBy("category") \
    .agg(avg("rating_performance").alias("Avg_Performance")) \
    .filter(col("Avg_Performance").isNotNull()) \
    .orderBy("Avg_Performance", ascending=False)

display(final_summary)
```
